```python
from backbones.gmdit import GMDiT

# 사용 예
model = GMDiT()
print(model)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

def display_class_images(pixel_samples, class_ids=None, figsize_per_image=1.5, display_title=""):
    samples = list(pixel_samples)
    processed = []
    for img in samples:
        if isinstance(img, Image.Image):
            arr = np.array(img)
        else:
            arr = np.array(img)
            if arr.ndim == 3 and arr.shape[0] in (1,3):
                arr = arr.transpose(1, 2, 0)
        processed.append(arr)

    n = len(processed)
    if class_ids is None:
        class_ids = list(range(n))

    fig, axes = plt.subplots(1, n, figsize=(n * figsize_per_image, figsize_per_image))
    if n == 1:
        axes = [axes]

    # 전체 제목 추가
    if display_title:
        fig.suptitle(display_title, fontsize=10, y=1.1)  # y 값을 1.02로 올려서 제목을 위로 이동
    # 플롯 레이아웃 조정: top을 낮춰서 제목과 플롯 사이에 공간 확보
    
    for ax, img_arr, lbl in zip(axes, processed, class_ids):
        ax.imshow(img_arr)
        ax.axis("off")
        ax.set_title(f"Class {lbl}", fontsize=8)

    plt.show()
```

```python
from solvers.dual.gdual_box_solver import GDual_Box_Solver

# 클래스 0–9로 설정
class_ids = [i for i in range(0, 100, 10)]
model_fn, noise_schedule, noises = model.get_model_fn(pos_conds=class_ids, guidance_scale=3.0)
solver = GDual_Box_Solver(model_fn, noise_schedule, algorithm_type="dual_prediction")

latents = solver.sample(noises, steps=5, skip_type="time_uniform_flow", flow_shift=3.0, gamma=1.0, tau_x=0.0, tau_e=0.0)
samples = model.decode_vae(latents)
display_class_images(samples, class_ids, display_title='Box')

latents = solver.sample(noises, steps=5, skip_type="time_uniform_flow", flow_shift=3.0, gamma=1.0, tau_x=0.5, tau_e=0.0)
samples = model.decode_vae(latents)
display_class_images(samples, class_ids, display_title='Box')

latents = solver.sample(noises, steps=5, skip_type="time_uniform_flow", flow_shift=3.0, gamma=1.0, tau_x=1.0, tau_e=0.0)
samples = model.decode_vae(latents)
display_class_images(samples, class_ids, display_title='Box')

latents = solver.sample(noises, steps=5, skip_type="time_uniform_flow", flow_shift=3.0, gamma=1.0, tau_x=1.5, tau_e=0.0)
samples = model.decode_vae(latents)
display_class_images(samples, class_ids, display_title='Box')
```
